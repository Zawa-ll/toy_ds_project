```r
library("tidyverse") 
```

```r
 diamond_df <- read_csv("data/Diamonds Prices2022.csv")
diamond_df 
```

```r
diamond_df_filtered <- diamond_df |>
    select(-x, -y, -z, -table, -depth, -color, -`...1`)
diamond_small <- diamond_df_filtered |>
    filter(price < 1000 & price > 0)
diamond_middle <- diamond_df_filtered |>
    filter(price < 2000 & price >= 1000)
diamond_high <- diamond_df_filtered |>
    filter(price >= 3000)



# ,
#           medium_price = low
    #mutate(test = filter(price > 0 & price < 1000))
# pivot_wider(names_from = )


diamond_small <- diamond_small |> mutate(price = "L")
# diamond_small
diamond_middle <- diamond_middle |> mutate(price = "M")
# diamond_middle
diamond_high <- diamond_middle |> mutate(price = "H")
# diamond_high

new <- diamond_small |>
bind_cols(diamond_middle)

new
```
